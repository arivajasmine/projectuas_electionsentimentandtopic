```python
# Read dataset
import pandas as pd
df = pd.read_csv("traindata.csv")
df.head()
```

```python
import re
def remove_emoji(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"
                               u"\U0001F300-\U0001F5FF"
                               u"\U0001F680-\U0001F6FF"
                               u"\U0001F700-\U0001F77F"
                               u"\U0001F780-\U0001F7FF"
                               u"\U0001F800-\U0001F8FF"
                               u"\U0001F900-\U0001F9FF"
                               u"\U0001FA00-\U0001FA6F"
                               u"\U0001FA70-\U0001FAFF"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

def prepro(text):
    text = re.sub(r'[#@]\w+', '', text)  # Hapus hashtags dan mention @
    text = re.sub(r'[^\w\s]', '', text)  # Hapus tanda baca
    text = re.sub(r"\d+", "", text)  # Hapus teks numerik
    text = text.lower()  # Ubah jadi lower case
    text = re.sub(r' +', ' ', text).strip()  # Beresin spasi
    text = re.sub(r'http\S+', '', text)  # Hapus http
    text = remove_emoji(text)  # Hapus emoji
    return text

df["clean"] = df["full_text"].map(prepro)
df.head()
```

```python
def remove_stopwords(text):
    # Hapus stopwords dan hapus text yang hanya terdiri dari 3 kata atau kurang (biasanya singkatan)
    text = " ".join(word for word in text.split() if word not in stopwords and len(word) > 3)
    return text
```

### IndoBERT (Transformer)

```python
from transformers import AutoTokenizer, AutoModelForMaskedLM

tokenizer = AutoTokenizer.from_pretrained('indolem/indobert-base-uncased')
model = AutoModelForMaskedLM.from_pretrained("indolem/indobert-base-uncased").to("cuda")
```

```python
!nvidia-smi
```

```python
from tqdm import tqdm

embeddings = []
for tweet in tqdm(df["clean"]):
    encoded_input = tokenizer(tweet, return_tensors='pt', max_length=200, padding=True, truncation=True).to("cuda")
    output = model(**encoded_input).logits.to("cpu")
    embeddings.append(output)
```
